# Header


There are 3 types of cells:
 - code cells, for R code
 - markdown cells, for nice text
 - raw cells, for text

## Code cells allow you to enter and run code



Double click or use `Enter` to edit a cell, use `Esc` to exit and enter what is called *command mode*

Run a code cell using `Shift-Enter` or pressing the <button class='jp-ToolbarButtonComponent'><span class="jp-RunIcon jp-Icon jp-Icon-16 jp-ToolbarButtonComponent-icon"/></button> button in the toolbar above:

```r
weight <- 60
print(weight)
```

**Always run the code** you see in these notebooks. 

There are two other keyboard shortcuts for running code:

* `Alt-Enter` runs the current cell and inserts a new one below.
* `Ctrl-Enter` run the current cell and enters command mode.

#### Cells

Create new cells using the keys `a` (for *above*) and `b` (for *below*) in command mode or use the <span class="jp-AddIcon jp-Icon jp-Icon-16 jp-ToolbarButtonComponent-icon"></span> button.

Move around the cells with the `arrow up` `arrow down` keys or with `j` and `k`

You can delete cell by double pressing `d`

Change the type of a cell with the dropdown below or with:
- `m` for markdown
- `y` for code
- `r` for raw text
